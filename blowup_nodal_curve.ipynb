```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definindo a função da curva nodal
def nodal_curve(x):
    return np.sqrt(x**2 + x**3)

# Definindo a função phi(x,y) = (x,y,y/x)
def phi(x, y):
    return (x, y, y/x)

# Valores de x para a plotagem
x = np.linspace(-2, 2, 400)
y_positive = nodal_curve(x)
y_negative = -nodal_curve(x)

# Filtrando os pontos onde x != 0 para evitar divisão por zero
mask = x != 0
x_filtered = x[mask]
y_positive_filtered = y_positive[mask]
y_negative_filtered = y_negative[mask]

# Aplicando a função phi nos pontos da curva
phi_points_positive = np.array([phi(xi, yi) for xi, yi in zip(x_filtered, y_positive_filtered)])
phi_points_negative = np.array([phi(xi, yi) for xi, yi in zip(x_filtered, y_negative_filtered)])

# Plotando a curva nodal no plano 2D
plt.figure(figsize=(10, 5))
plt.plot(x, y_positive, label=r'$y^2 = x^2 + x^3$ (positivo)')
plt.plot(x, y_negative, label=r'$y^2 = x^2 + x^3$ (negativo)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Curva Nodal no Plano 2D')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.show()

# Plotando a curva mapeada no espaço 3D
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111, projection='3d')
ax.plot(phi_points_positive[:, 0], phi_points_positive[:, 1], phi_points_positive[:, 2], label=r'$\phi(x,y)$ para $y > 0$')
ax.plot(phi_points_negative[:, 0], phi_points_negative[:, 1], phi_points_negative[:, 2], label=r'$\phi(x,y)$ para $y < 0$')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z = y/x')
ax.legend()
ax.set_title('Curva Mapeada no Espaço 3D pela Função φ')
plt.show()
```
